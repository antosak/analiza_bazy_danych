```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Create days list
col_num = [0] + list(range(1, 63, 2))
col_names = ['name']
for i in range(1, 10):
    col_names.extend(['0{}'.format(i)])
for i in range(10, 32):
    col_names.extend(['{}'.format(i)])
    
# Read original data
df = pd.read_excel('../Analysis Data/weather_excel.xlsx', index_col=None, header=None, names=col_names, usecols=col_num)
df.replace(-9999, pd.NA, inplace=True)


#Separation of information in the first column
for i in range(0, df.count()[0]):
    first_col = df['name'][i]
    new_col = first_col[11:15] + ' ' + first_col[15:17] + ' ' + first_col[17:21]
    df['name'][i] = new_col

df[['year', 'month', 'kind']] = df.name.str.split(' ', expand=True)
df.drop('name', axis=1, inplace=True)

# Drop useless records
df = df[df['year'] == '2010']
df = df[df['month'] <= '05']

df = df.melt(id_vars=['year', 'month', 'kind'])

df = df[(pd.notna(df['value']))]
df.set_axis(list(range(df.count()[0])), axis=0, inplace=True)
df['day'] = df['variable']
df.drop('variable', axis=1)
df = df[['year', 'month', 'day', 'kind', 'value']]
df.sort_values(by=["year", 'month', 'day'], inplace=True)
for index, row in df.iterrows():
    if row['value'] == 0:
        df.drop(index, inplace=True)

df['value'] = df['value'].div(10)
df.to_excel(excel_writer="../Analysis Data/tidy_weather.xlsx", index=False)
```
